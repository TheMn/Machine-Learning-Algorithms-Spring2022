```python
from sklearn import datasets

circles = datasets.make_circles(n_samples=500, factor=0.5, noise=0.1)
blobs = datasets.make_blobs(n_samples=500, random_state=8)

circles_X, circles_y = circles[0], circles[1]
blobs_X, blobs_y = blobs[0], blobs[1]
```

```python
import matplotlib.pyplot as plt 

plt.scatter(circles_X[:, 0], circles_X[:, 1], c=circles_y)
```

```python
import scipy.cluster.hierarchy as shc

shc.dendrogram(shc.linkage(circles_X, 'ward'),
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
import scipy.cluster.hierarchy as shc

shc.dendrogram(shc.linkage(circles_X, 'complete'),
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='complete')
model.fit_predict(circles_X)
```

```python
plt.scatter(circles_X[:,0], circles_X[:,1], c=model.labels_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(circles_X, model.fit_predict(circles_X))
```

```python
plt.scatter(blobs_X[:, 0], blobs_X[:, 1], c=blobs_y)
```

```python
shc.dendrogram(shc.linkage(blobs_X, 'ward'),
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit_predict(blobs_X)
```

```python
plt.scatter(blobs_X[:,0], blobs_X[:,1], c=model.labels_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(blobs_X, model.fit_predict(blobs_X))
```
