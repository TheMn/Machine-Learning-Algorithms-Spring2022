```python
from sklearn.datasets import load_breast_cancer

X, y = load_breast_cancer(return_X_y=True)
X.shape
```

```python
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(kernel="cosine", fit_inverse_transform=True, gamma=5, n_components=2)
X_kpca = kpca.fit_transform(X)
X_kpca.shape
```

```python
import pandas as pd

kpca_res = pd.DataFrame(X_kpca)
kpca_res.columns=["kpca1","kpca2"]
kpca_res.head()
kpca_res['y']=y
```

```python
import matplotlib.pyplot as plt

plt.scatter(kpca_res['kpca1'].values, kpca_res['kpca2'].values, c=kpca_res['y'].values)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(kpca_res, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0).fit(X_train, y_train)
clf.score(X_test, y_test)
```
