```python
from sklearn import datasets

circles = datasets.make_circles(n_samples=500, factor=0.5, noise=0.1)
blobs = datasets.make_blobs(n_samples=500, random_state=8)

circles_X, circles_y = circles[0], circles[1]
blobs_X, blobs_y = blobs[0], blobs[1]
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.12, min_samples=3).fit(circles_X)
labels = db.labels_
```

```python
import matplotlib.pyplot as plt
plt.scatter(circles_X[:, 0], circles_X[:, 1], c=labels)
```

```python
import numpy as np
np.unique(labels).size
```

```python
np.sum(np.array(labels) == -1, axis=0)
```

```python
plt.scatter(circles_X[:, 0], circles_X[:, 1], c=['g' if l!=-1 else 'r' for l in labels])
```
