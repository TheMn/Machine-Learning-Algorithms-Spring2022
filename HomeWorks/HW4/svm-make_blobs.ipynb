```python
from sklearn.datasets import make_blobs
data, labels = make_blobs(n_samples=[150]*4, n_features=3, random_state=142)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data[:, 0], data[:, 1], linewidths=data[:, 2], c=labels)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
from sklearn import svm

linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X_train, y_train)
rbf = svm.SVC(kernel='rbf', gamma=1, C=1, decision_function_shape='ovo').fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1, decision_function_shape='ovo').fit(X_train, y_train)
sig = svm.SVC(kernel='sigmoid', C=1, decision_function_shape='ovo').fit(X_train, y_train)
```

```python
linear.predict(X_test)
```

```python
import numpy as np
plt.scatter(X_test[:, 0], X_test[:, 1], linewidths=X_test[:, 2], c=np.subtract(y_test, sig.predict(X_test)))
```

```python
print("Accuracy Linear Kernel:", linear.score(X_test, y_test))
print("Accuracy Polynomial Kernel:", poly.score(X_test, y_test))
print("Accuracy Radial Basis Kernel:", rbf.score(X_test, y_test))
print("Accuracy Sigmoid Kernel:", sig.score(X_test, y_test))
```
